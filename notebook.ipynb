```python
newdata = data.copy()
newdata['Age at First Oscar'] = newdata['Year of First Oscar/Grammy/Emmy'] - newdata['Birth Year']

# Drop rows where 'Year of first oscar' or 'Birth year' is missing
newdata.dropna(subset=['Year of First Oscar/Grammy/Emmy', 'Birth Year'], inplace=True)
breakthrough_counts = newdata.groupby('Age at First Oscar').size()

plt.figure(figsize=(10, 6))
breakthrough_counts.plot(kind='line', marker='o', color='blue')
plt.title('Distribution of Age at First Oscar')
plt.xlabel('Age at First Oscar')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from lifelines import KaplanMeierFitter

newdata = data.copy()
newdata.dropna(subset=['Year of Death'], inplace=True)

# Calculate the duration from birth to death
newdata['Duration'] = newdata['Year of Death'] - newdata['Birth Year']

# Initialize KaplanMeierFitter
kmf = KaplanMeierFitter()

# Fit the Kaplan-Meier estimator
kmf.fit(newdata['Duration'], event_observed=(newdata['Duration'] < newdata['Duration'].max()))

# Plot the survival curve
plt.figure(figsize=(10, 6))
kmf.plot()
plt.title('Survival Curve for Entertainers')
plt.xlabel('Years')
plt.ylabel('Survival Probability')
plt.grid(True)
plt.show()
```

```python
print(len(newdata))
print(len(data))
```

```python
newdata = data.copy()

newdata.dropna(subset=['Year of Breakthrough/#1 Hit/Award Nomination', 'Year of Last Major Work (arguable)'], inplace=True)

# Calculate the duration between breakthrough and last major work
newdata['Years Between Breakthrough and Last Major Work'] = newdata['Year of Last Major Work (arguable)'] - newdata['Year of Breakthrough/#1 Hit/Award Nomination']
breakthrough_counts = newdata.groupby('Years Between Breakthrough and Last Major Work').size()

plt.figure(figsize=(10, 6))
breakthrough_counts.plot(kind='line', marker='o', color='blue')
plt.title('Distribution of Years Between Breakthrough and Last Major Work')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
